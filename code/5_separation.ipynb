```python
import pandas as pd
```

```python
# separate student, parent, and teacher surveys
sur_19 = pd.read_csv('../data/survey_scores_clean_19.csv')
```

```python
sur_19.head(2)
```

```python
# getting student survey results only (+ scores)
st_cols = ['dbn', 'school_name', 'student_response_rate'] # the first 3 column names

for column in sur_19.columns:
    if column.endswith('s'):
        st_cols.append(column)
    
# append the scores
st_cols.append('combined_mean_score')

# get the data from only the columns in st_cols    
st_19 = sur_19[st_cols]
```

```python
# getting parent survey results only (+ scores)
pa_cols = ['dbn', 'school_name', 'parent_response_rate'] # the first 3 column names

for column in sur_19.columns:
    if column.endswith('p'):
        pa_cols.append(column)
    
# append the scores
pa_cols.append('combined_mean_score')

# get the data from only the columns in st_cols    
pa_19 = sur_19[pa_cols]
```

```python
pa_19.head(2)
```

```python
# getting teacher survey results only (+ scores)
te_cols = ['dbn', 'school_name', 'parent_response_rate'] # the first 3 column names

for column in sur_19.columns:
    if column.endswith('t'):
        te_cols.append(column)
    
# append the scores
te_cols.append('combined_mean_score')

# get the data from only the columns in st_cols    
te_19 = sur_19[te_cols]
```

```python
te_19.head(2)
```

```python
st_19.to_csv('../data/2019_student_ready.csv', index=False)
pa_19.to_csv('../data/2019_parent_ready.csv', index=False)
te_19.to_csv('../data/2019_teacher_ready.csv', index=False)
```
